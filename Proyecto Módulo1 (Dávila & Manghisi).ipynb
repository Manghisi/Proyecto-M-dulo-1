* Obvervar que polinomio se ajusta mejor a la curva y graficar el error cuadrático

### Modelos que representan el problema

Con ayudad de un potenciometro (medidor de pH), se realizó una titulación de cerveza. Como base se utilizó hidróxido de sodio (NaOH).

Se titularon 15[mL] de cerveza a la cuál se le añadieron de manera gradual 25[mL] de NaOH a 0.1M. Se registro una lectura del potencial de hudrógeno cada 0.5[mL] agregados de NaOH.

Se realizo una tabla donde se agregaron todos los datos de las mediciones de pH de la solución en función del volumen de NaOh añadido.

```python
import pandas as pd
file_name = 'Datos Proyecto Módulo1 (Dávila & Manghisi).csv'
datos = pd.read_csv(file_name)
datos
```

```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
%matplotlib inline

x = datos.iloc[:,1].values
y = datos.iloc[:,2].values

def funcion (beta,x,n):
    return beta@[x**i for i in range(n+1)]

def objetivo (beta, x, y, n):
    return np.sum((y-funcion(beta,x,n))**2)

ajuste_1 = opt.minimize(objetivo,np.ones((2,)),args=(x,y,1))
ajuste_2 = opt.minimize(objetivo,np.ones((3,)),args=(x,y,2))
ajuste_3 = opt.minimize(objetivo,np.ones((4,)),args=(x,y,3))
ajuste_4 = opt.minimize(objetivo,np.ones((5,)),args=(x,y,4))
ajuste_5 = opt.minimize(objetivo,np.ones((6,)),args=(x,y,5))
ajuste_6 = opt.minimize(objetivo,np.ones((7,)),args=(x,y,6))
ajuste_7 = opt.minimize(objetivo,np.ones((8,)),args=(x,y,7))
```

```python
plt.plot(x,y,'og',label='Datos')
plt.xlabel('Volumen de NaOH')
plt.ylabel('pH de la cerveza')
plt.legend(loc='best')
plt.grid()
```

```python
plt.plot(x,y,'og',ms=3,label='Datos')
plt.plot(x,funcion(ajuste_1.x,x,1),'b',label='Ajuste grado 1')
plt.grid()
plt.xlabel('Volumen de NaOH')
plt.ylabel('pH de la cerveza')
plt.legend(loc='best')
```
